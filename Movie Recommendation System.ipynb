**Importing the Library**

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import  TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Preprocessing

```python
movies_data = pd.read_csv(r"movies.csv")
```

```python
# printing first 5 rows of dataframe
movies_data.head()
```

```python
# number of rows and columns in dataframe 
movies_data.shape
```

```python
# Selecting the relevant features for recommendation
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
movies_data.isnull().sum()
```

```python
# replacing the null value with null string

for features in selected_features:
    movies_data[features] = movies_data[features].fillna('')
```

```python
# combining all the selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
# combined_features=pd.DataFrame()
# for features in selected_features:
#     combined_features[features]=movies_data[features]
combined_features
```

```python
# converting text data to features vectors
vectorizer=TfidfVectorizer()
features_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(features_vectors)
```

Consine Similarity

```python
# getting similarity scores using consine similarity
similarity = cosine_similarity(features_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
# getting the movie name from the user
movie_name = input("Enter your favourite movie name : ")
```

```python
# creating a list with all the movie names given in the movie name given in the dataset
list_of_all_title = movies_data['title'].tolist()
print(list_of_all_title)
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_title)
print(find_close_match)
if(find_close_match==[]):
    print("Movie Not Found......")
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of movie with title
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting the list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies=sorted(similarity_score,key= lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index
print("Movies suggested for you : \n")
i=1
for movies in sorted_similar_movies:
    index = movies[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<=30):
        print(i,'.',title_from_index)
        i+=1
```
